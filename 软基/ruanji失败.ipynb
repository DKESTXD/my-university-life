```python
#从所有人脸集中筛选出含有至少20张图片的人物作为数据集，共有62个人
import os
root=".\lfw\lfw"#根目录
datas_list=[]
image_extensions=('.jpg','.jpeg', '.png', '.gif', '.bmp', '.tiff')
for folder_name in os.listdir(root):
    folder_path = os.path.join(root, folder_name)
    if os.path.isdir(folder_path):
        image_count=0
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)            
            if os.path.isfile(file_path) and file_name.lower().endswith(image_extensions):
                image_count += 1
        if image_count >= 20:
            datas_list.append(folder_name)

print(datas_list)
print(len(datas_list))
```

```python
#存储所有图片路径的二维列表
image_paths = []
for diffname in datas_list:
    diffpath = os.path.join(root, diffname)
    if os.path.isdir(diffpath):
        files = os.listdir(diffpath)
        image_files = [os.path.join(diffpath, file) for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]
        image_paths.append(image_files)
#每个人都只保留20张图片
import random
data_paths=[]
for lists in image_paths:
    new_lists=random.sample(lists,10)
    data_paths.append(new_lists)
print(data_paths)
```

```python
#将图片路径加载成图片得到图片列表datas
import cv2
datas=[]
for paths in data_paths:
    images = []  # 存储当前子列表中的图片对象
    for path in paths:
        img = cv2.imread(path)
        images.append(img)
    datas.append(images) 
print(datas)
```

```python
#图片预处理
import numpy as np
datas_vector=[]
for images in datas:
    image_vectors=[]
    for img in images:
        img=cv2.resize(img,(150,150))#统一大小
        img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#转化为灰度图
        img=img.flatten()#展平为一维向量
        image_vectors.append(img)
    datas_vector.extend(image_vectors)#按列整合为矩阵
for i in datas_vector:
    print(i)
```

```python
#转化为numpy数组便于矩阵运算
X=np.array(datas_vector)
Y=np.array(datas_list)
print(X)
print(X.shape)
```

```python
#建立图像数据与标签的索引关系
X_index=np.array([])
for i in range(0,X.shape[0]):
    X_index=np.append(X_index,i//10)
print(X_index)
```

```python
#划分训练集与测试集
X_column=[range(0,620)]
X_train_index=[]
for i in range(0,62):
    X_part=np.array(range(10*i,10*i+10))
    X_train_part=np.random.choice(X_part,size=7,replace=False,p=None)#随机数取0.7的数据作为训练集
    X_train_index.extend(X_train_part)
X_train_index.sort()
X_test_index=np.setdiff1d(X_column,X_train_index)#余下的0.3数据即为测试集
X_train_index=np.setdiff1d(X_column,X_test_index)
print(X_train_index)#训练集的索引
print(X_test_index)#测试集的索引
```

```python
#根据索引划分出训练集和测试集
#这样测试时可以根据其列序号得到对应的标签序号
X_train=X[X_train_index,:]
X_test=X[X_test_index,:]
print(X_train)
print(X_test)
```

```python
#求差值矩阵的协方差矩阵
X_cov=np.cov(X_diff)#求协方差矩阵
print(X_cov.shape)
#求协方差矩阵的特征值和特征向量
featvalue,featvector=np.linalg.eig(X_cov)
print(featvector.shape)
#得到了特征的排序，取前400个最有影响的特征
feature=featvector[:,0:40]
```

```python
print(feature.shape)
X_feature=np.dot(X_diff.T,feature)
for i in range(0,40):
    X_feature[:,i]=X_feature[:,i]/np.linalg.norm(X_feature[:,i])
X_final=np.dot(X_diff,X_feature)
print(X_final)
#得到样本差值矩阵在特征空间上的投影
```
